```python
scraper = ygfandom.WebScraper()
```

```python
scraper.set_card_urls('https://yugioh.fandom.com/wiki/Eternity_Code')
scraper.get_card_urls()
```

```python
cards_url_list = scraper.get_card_urls()
```

```python
time1 = time.perf_counter()
for link in cards_url_list:
    scraper.set_card_details(link)

print(f'{time.perf_counter() - time1} seconds')
scraper.get_card_details()
```

```python
print(len(scraper.get_card_details()))
```

```python
scraper2 = ygfandom.WebScraper()
scraper2.set_card_urls('https://yugioh.fandom.com/wiki/Eternity_Code')
cards_url_list = scraper2.get_card_urls()
print(len(cards_url_list))
```

```python
time2 = time.perf_counter()

if __name__ == "__main__":
    with concurrent.futures.ThreadPoolExecutor(max_workers = 5) as executor:
        executor.map(scraper2.set_card_details, cards_url_list)
        
    print(f'{time.perf_counter() - time2} seconds')
```

```python
scraper2.get_card_details()
```
